```python
from sklearn import preprocessing
```

```python
#le = preprocessing.LabelEncoder().fit(np.round(lsamples))
```

```python
#blsamples = np.where(lsamples < 500, 0,1 )
std_samples =preprocessing.StandardScaler().fit_transform(samples)#.iloc[:,0:8])
```

```python
#X_train, X_test, y_train, y_test = train_test_split(cuttedsamples,lsamples, test_size=0.2, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(std_samples,lsamples, test_size=0.2, random_state=0)
```

```python
#y_train = le.transform(y_train)
#y_test = le.transform(y_test)
```

```python
#X_train =preprocessing.StandardScaler().fit_transform(X_train)
#X_test = preprocessing.StandardScaler().fit_transform(X_test)
DecisionTreeRegressor?
```

```python
regressor = DecisionTreeRegressor(splitter='best', max_features="sqrt",min_samples_split=5, max_depth=10)#,min_impurity_decrease=0.8)
```

```python
regressor.fit(X_train, y_train) #training the algorithm
```

```python
y_pred = regressor.predict(X_test)
```

```python
#plt.scatter(X_test, y_test,  color='gray')
plt.figure()
plt.title("Prediction Map")
plt.scatter(X_test[:,0], X_test[:,1], c=y_pred, cmap='cool')
plt.figure()
plt.title("Labeled Map")
plt.scatter(X_test[:,0], X_test[:,1], c=y_test, cmap='cool')
plt.show()
```

```python
#print('Accuracy:', metrics.accuracy_score(y_test, y_pred))  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Median Absolue Error:', metrics.median_absolute_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
df2 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df3 = df2.head(80)
df3.plot(kind='bar',figsize=(15,5))
plt.grid(which='major', linestyle='-', linewidth='0.3', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.3', color='black')
plt.show()
```

```python
regressor.get_n_leaves(), regressor.get_depth(), regressor.get_params()
```

```python
from sklearn.tree import plot_tree
plt.figure()
plot_tree(regressor, filled=True)
plt.savefig("/media/WIP/tree.png")
plt.show()
```
