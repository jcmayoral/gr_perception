Linear Regression
https://towardsdatascience.com/a-beginners-guide-to-linear-regression-in-python-with-scikit-learn-83a8f7ae2b4f

```python
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVR
from sklearn import metrics
import seaborn as seabornInstance 
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
seabornInstance.distplot(labels)
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
seabornInstance.distplot(df['MeanX'])
plt.figure(figsize=(10,5))
plt.tight_layout()
seabornInstance.distplot(df['MreanY'])
plt.figure(figsize=(10,5))
plt.tight_layout()
seabornInstance.distplot(df[' MeanZ'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(samples, lsamples, test_size=0.2, random_state=0)
```

```python
regressor = SVR(kernel='rbf', C=1000, gamma=0.01, epsilon=.01)
#(verbose=2)  
```

```python
#potential overfitting
regressor.fit(X_train, y_train) #training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
```

```python
y_pred = regressor.predict(X_test)
```

```python
df2 = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df2
```

```python
testcenters = X_test.to_numpy()
predictedlabels = y_pred
```

```python
plt.scatter(testcenters[:,0], testcenters[:,1], c=predictedlabels, cmap='hot')
```

```python
df3 = df2.head(25)
df3.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
#plt.scatter(X_test, y_test,  color='gray')
plt.figure()
plt.title("Prediction Map")
plt.scatter(testcenters[:,0], testcenters[:,1], c=predictedlabels, cmap='hot')
plt.figure()
plt.title("Labeled Map")
plt.scatter(testcenters[:,0], testcenters[:,1], c=y_test, cmap='hot')
plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
